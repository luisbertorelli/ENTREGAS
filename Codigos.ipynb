```python
import tkinter as tk

# Definir productos y sus precios
productos = {
    1: {"nombre": "Chicles", "precio": 1000},
    2: {"nombre": "Chocorramo", "precio": 1500},
    3: {"nombre": "Papas BBQ", "precio": 2000},
    4: {"nombre": "Agua", "precio": 2500},
    5: {"nombre": "Jugo HIT Naranja Piña", "precio": 3000},
    6: {"nombre": "Mr Tea Limón", "precio": 3500}
}

# Definir billetes aceptados
billetes_aceptados = [10000, 5000, 2000]

# Función para mostrar productos
def mostrar_productos():
    for id, producto in productos.items():
        print(f"{id}. {producto['nombre']} - Precio: {producto['precio']} pesos")

# Función para realizar la compra
def comprar_producto(producto_id, monto_insertado):
    if producto_id in productos:
        producto = productos[producto_id]
        precio = producto['precio']
        if monto_insertado >= precio:
            cambio = monto_insertado - precio
            print(f"Compra exitosa. Recoge tu {producto['nombre']}. Cambio: {cambio} pesos")
        else:
            print("Saldo insuficiente. Ingresa un monto mayor o igual al precio del producto.")
    else:
        print("Producto no válido.")

def crear_interfaz():
    # Crear ventana
    ventana = tk.Tk()
    ventana.title("Máquina Dispensadora")

    # Función para realizar la compra
    def comprar():
        producto_id = seleccion.get()
        if producto_id in productos:
            monto_insertado = int(monto_entry.get())
            if monto_insertado in billetes_aceptados:
                comprar_producto(producto_id, monto_insertado)
                resultado.set(f"Compra exitosa. Recoge tu {productos[producto_id]['nombre']}. Cambio: {monto_insertado - productos[producto_id]['precio']} pesos")
            else:
                resultado.set("Billete no válido.")
        else:
            resultado.set("Producto no válido.")

    # Mostrar productos
    mostrar_productos()

    # Crear variable para almacenar la selección
    seleccion = tk.IntVar()
    seleccion.set(1)

    # Crear etiqueta de selección de producto
    etiqueta_seleccion = tk.Label(ventana, text="Selecciona un producto:")
    etiqueta_seleccion.pack()

    # Crear botones para seleccionar productos
    for id, producto in productos.items():
        tk.Radiobutton(ventana, text=f"{producto['nombre']} - {producto['precio']} pesos",
                       variable=seleccion, value=id).pack(anchor=tk.W)

    # Crear etiqueta y entrada para ingresar monto
    etiqueta_monto = tk.Label(ventana, text="Ingresa el billete (2000, 5000 o 10000):")
    etiqueta_monto.pack()

    monto_entry = tk.Entry(ventana)
    monto_entry.pack()

    # Crear botón para comprar
    boton_comprar = tk.Button(ventana, text="Comprar", command=comprar)
    boton_comprar.pack()

    # Variable para mostrar el resultado
    resultado = tk.StringVar()
    resultado.set("")
    etiqueta_resultado = tk.Label(ventana, textvariable=resultado)
    etiqueta_resultado.pack()

    # Iniciar la ventana
    ventana.mainloop()

# Ejecutar el programa
if __name__ == "__main__":
    crear_interfaz()
```

```python
import hashlib

# El usuario proporciona la contraseña actual y la nueva contraseña
contraseña_actual = input("Ingresa tu contraseña actual: ")
nueva_contraseña = input("Ingresa tu nueva contraseña (entre 8 y 12 caracteres, con al menos una letra mayúscula): ")

def hash_contraseña(contraseña):
    return hashlib.sha256(contraseña.encode('utf-8')).hexdigest()

def verificar_contraseña_actual(contraseña_actual, hashed_password):
    return hash_contraseña(contraseña_actual) == hashed_password

# Simulación: Supongamos que 'hashed_password' es la contraseña almacenada previamente
hashed_password = hash_contraseña(contraseña_actual)

if not verificar_contraseña_actual(contraseña_actual, hashed_password):
    print("La contraseña actual es incorrecta")
elif hash_contraseña(nueva_contraseña) == hash_contraseña(contraseña_actual):
    print("La nueva contraseña debe ser diferente a la actual")
elif not (8 <= len(nueva_contraseña) <= 12):
    print("La nueva contraseña debe tener entre 8 y 12 caracteres")
elif not any(char.isupper() for char in nueva_contraseña):
    print("La nueva contraseña debe contener al menos una letra mayúscula")
else:
    print("Contraseña cambiada exitosamente")

```
